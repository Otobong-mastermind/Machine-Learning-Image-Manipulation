# Converting the Images above across different spaces takes the same route
#You use the color.name-of-colorspace(original img) to generate and save

```python
#To save an Image
from skimage import io
from skimage import color
from skimage import data
from pylab import *
img_rgb = io.imread('City-Image.jpg')
#Convert to YPbPr
img_ypbpr = color.rgb2ypbpr(img_rgb)
#Display
io.imshow(img_ypbpr)
#Save the Image
io.imsave("newypbpr.jpg",img_ypbpr)
```

# Drawing a Line Over an Image

```python
from skimage import io
from skimage import draw
img = io.imread('City-image.jpg')
x,y = draw.line(0,0,1500,1500)
img[x,y] = 0
io.imshow(img)
```

```python
from skimage import io
from skimage import draw
img = io.imread('City-image.jpg')
def rectangle(w,x,y,h):
    rr,cc = [x,x+w,x+w,x],[y,y,y+h,y+h]
    return(draw.polygon(rr,cc))
rr,cc = rectangle(500,500,500,500)
img[rr,cc] = 1
io.imshow(img)
```

# Drawing a Circle

```python
from skimage import io
from skimage import draw
img = io.imread('City-image.jpg')
x,y = draw.circle(1000,1000,500)
img[x,y] = 1
io.imshow(img)
```

# Drawing a Bezier Curve

```python
from skimage import io
from skimage import draw

img = io.imread('City-image.jpg')
(x,y) = draw.bezier_curve(0,0, 500, 500, 900,1200,100)

img[x,y] = 1
io.imshow(img)
```

# Doing Gamma Correction on an image

```python
from skimage import io
from skimage import exposure
from pylab import *
img = io.imread('City-image.jpg')
#To work on the gamma
new_image1 = exposure.adjust_gamma(img, 0.5)
new_image2 = exposure.adjust_gamma(img, 3)
figure(0)
io.imshow(new_image1)
figure(1)
io.imshow(new_image2)
```

# Transforming Images

```python
#This code example Rotates the image

from skimage import io
from skimage.transform import rotate
img = io.imread('City-image.jpg')
rot_img = rotate(img,20)
io.imshow(rot_img)
```

```python
#This code example Resizes the image

from skimage import io
from skimage.transform import resize
img = io.imread('City-image.jpg')
rez_img = resize(img,(1200,2000))
io.imshow(rez_img)
```

# Determining the Structural Similarity Between Images

```python
from skimage import io
from skimage.measure import compare_ssim as ssim
img_original = io.imread('City-image.jpg')
img_modified = io.imread('newypbpr.jpg')
ssim_original = ssim(img_original, img_original, data_range=img_original.max() - img_original.min(), multichannel= True)
ssim_different = ssim(img_original, img_modified, data_range=img_modified.max() - img_modified.min(), multichannel= True)
print(ssim_original,ssim_different)
```
